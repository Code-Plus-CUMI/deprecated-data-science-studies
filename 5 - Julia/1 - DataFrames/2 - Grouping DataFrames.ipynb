<b>1) Reading the Dataset</b>

```julia
# Environment Variables
ENV["LINES"], ENV["COLUMNS"] = 15, 200
```

```julia
# Reading the DataSet (CSV)
df = CSV.File("auto2.csv") |> DataFrame
```

```julia
# Reading the DataSet (Arrow)
df2 = Arrow.Table("auto2.arrow") |> DataFrame
```

----

<b>2) Grouping By</b>

```julia
# Simple grouping by "horsepower"
gdf = groupby(df, :horsepower)
```

```julia
# Calculating Mpg Mean
combine(gdf, :mpg => mean)
```

```julia
# Calculating Mpg Mean and Naming it as mpgMean
mean_gdf = combine(gdf, :mpg => mean => :mpgMean)
```

```julia
# Sorting by mpgMean (in ascending order)
sort!(mean_gdf, :mpgMean, rev=false)
```

----

<b>3) Frequence Tables</b>

```julia
# Frequence Table of one feature
freqtable(df, :mpg)
```

```julia
# Frequence Table of n features
freqtable(df, :mpg, :cylinders)
```

----

<b>4) Pipes</b>

```julia
# Grouping the dataframe by mpg feature
# Calculating its mean and naming it mpgMean
# and Sorting by mpgMean in descending order
new_df = @pipe df |>
               groupby(_, :mpg) |>
               combine(_, :mpg => mean => :mpgMean) |>
               sort(_, :mpgMean, rev=true)
```

```julia
# Now we will do the same thing but replacing all missing
# values per zeros
new_df = @pipe df |>
               coalesce.(_, 0) |>
               groupby(_, :mpg) |>
               combine(_, :mpg => mean => :mpgMean) |>
               sort(_, :mpgMean, rev=true)
```
