<h2>---- Plotting ----</h2>

````julia
#Pkg.add("Plots")
using Plots
````

````julia
#Pkg.add("PlotlyJS")
using PlotlyJS
````

````julia
# Creating fake datas
power_2(x) = x^2

x = -3:0.1:3
y = power_2.(x)
````

----

````julia
gr() # loading the backend plot
plot(x, y, label="line") # line plot
scatter!(x, y, label="points") # scatter plot. "!" allows you to concatenate two or more plots
````

````julia
plotlyjs() # loading another background plot motor
plot(x, y, label="line") # line plot
scatter!(x, y, label="points") # scatter plot. "!" allows you to concatenate two or more plots
````

````julia
plotlyjs() # loading another background plot motor
plot(x, y, label="line") # line plot
scatter!(x, y, label="points") # scatter plot. "!" allows you to concatenate two or more plots

xflip!() # reverses x-axis

xlabel!("My Label on X-Axis") # label on x-axis
ylabel!("My Label on Y-Axis") # label on y-axis
title!("My Title") # title
````
